```python
# Import Dependencies
import sqlite3
import pandas as pd
```

```python
# Import CSV file
df = pd.read_csv('countries-table.csv')
df.set_index('country', inplace=True)
df.head()
```

```python
# Connect to SQLite db
connection = sqlite3.connect('countries.sqlite')
```

```python
# Load csv to SQLite db
df.to_sql('countries', connection, if_exists='replace', index=True, index_label='country')
```

```python
# Confirm primary key
cursor = connection.execute('PRAGMA table_info(countries)')
columns = cursor.fetchall()

primary_key_column = None
for column in columns:
    if column[5] == 1:
        primary_key_column = column[1]
        break

if primary_key_column is not None:
    print(f'The primary key column is {primary_key_column}.')
else:
    print('The table does not have a primary key.')
```

```python
# Close connection
connection.close()
```
